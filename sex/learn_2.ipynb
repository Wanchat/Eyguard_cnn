```python
import matplotlib
matplotlib.use("Agg")
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from keras.preprocessing.image import img_to_array
from keras.utils import to_categorical
from lenet_sex import LeNet
from imutils import paths
import matplotlib.pyplot as plt
import numpy as np
import argparse
import random
import cv2
import os
```

```python
EPOCHS = 25
INIT_LR = 1e-3
BS = 32
```

```python
data = []
labels = []
```

```python
imagePaths = sorted(list(paths.list_images(r"D:\code_python\cnn\sex\images")))
random.seed(42)
random.shuffle(imagePaths)
```

```python
for imagePath in imagePaths:
    # load the image, pre-process it, and store it in the data list
    image = cv2.imread(imagePath)
    image = cv2.resize(image, (28, 28))
    image = img_to_array(image)
    data.append(image)

    # extract the class label from the image path and update the
    # labels list
    label = imagePath.split(os.path.sep)[-2]
    label = 1 if label == "woman" else 0
    labels.append(label)
```

```python
for i in labels:
    print(i)
```

```python
# scale the raw pixel intensities to the range [0, 1]
data = np.array(data, dtype="float") / 255.0
labels = np.array(labels)
labels
```

```python
# partition the data into training and testing splits using 75% of
# the data for training and the remaining 25% for testing
(trainX, testX, trainY, testY) = train_test_split(data,labels, test_size=0.25, random_state=42)
```

```python
# convert the labels from integers to vectors
trainY = to_categorical(trainY, num_classes=2)
testY = to_categorical(testY, num_classes=2)
```

```python
# construct the image generator for data augmentation
aug = ImageDataGenerator(rotation_range=6, width_shift_range=0.1,height_shift_range=0.1, shear_range=0.3, zoom_range=0.2, horizontal_flip=True, fill_mode="nearest")
```

```python
# initialize the model
print("[INFO] compiling model...")
model = LeNet.build(width=28, height=28, depth=3, classes=2)
opt = Adam(lr=INIT_LR, decay=INIT_LR / EPOCHS)
model.compile(loss="binary_crossentropy", optimizer=opt, metrics=["accuracy"])
```

```python
# train the network
print("[INFO] training network...")
H = model.fit_generator(aug.flow(trainX, trainY, batch_size=BS), validation_data=(testX, testY), steps_per_epoch=len(trainX) // BS, epochs=EPOCHS, verbose=1)
```

```python
# save the model to disk
print("[INFO] serializing network...")
model.save("jupyter_model_sex.model")
```
